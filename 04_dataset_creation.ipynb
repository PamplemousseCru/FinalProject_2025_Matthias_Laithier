```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
```

```python
export_dir = "./exports"
big_matrix_cleaned = pd.read_parquet(export_dir + "/big_matrix_cleaned.pq")
user_df = pd.read_parquet(export_dir + "/user_df.pq")
video_df = pd.read_parquet(export_dir + "/video_df.pq")
```

```python
big_matrix_cleaned = big_matrix_cleaned.drop(columns=["video_duration"])
big_matrix_cleaned['date'] = pd.to_datetime(big_matrix_cleaned['time']).dt.date
big_matrix_cleaned = big_matrix_cleaned.drop(columns=["time"])
big_matrix_cleaned = big_matrix_cleaned.merge(user_df, on='user_id', how='left')
video_df['date'] = pd.to_datetime(video_df['date']).dt.date
big_matrix_cleaned = big_matrix_cleaned.merge(video_df, on=['video_id', 'date'], how='left')
```

```python
user_feature_cols = [col for col in user_df.columns if col not in ('user_id')]
video_feature_cols = [col for col in video_df.columns if col not in ('video_id', 'date')]
scaler = StandardScaler()
X_user = scaler.fit_transform(big_matrix_cleaned[user_feature_cols])
print("yay")
scaler = StandardScaler()
X_video = scaler.fit_transform(big_matrix_cleaned[video_feature_cols])
y = pd.DataFrame(big_matrix_cleaned['watch_ratio'])
```

```python
user_feature_cols = [col for col in user_df.columns if col not in ('user_id')]
video_feature_cols = [col for col in video_df.columns if col not in ('video_id', 'date')]

scaler = StandardScaler()
X_user = pd.DataFrame(
    scaler.fit_transform(big_matrix_cleaned[user_feature_cols]),
    columns=user_feature_cols,
    index=big_matrix_cleaned.index
)
print("yay")
# Scale video features and preserve DataFrame format
scaler = StandardScaler()
X_video = pd.DataFrame(
    scaler.fit_transform(big_matrix_cleaned[video_feature_cols]),
    columns=video_feature_cols,
    index=big_matrix_cleaned.index
)

# Target
y = pd.DataFrame(big_matrix_cleaned['watch_ratio'])
```

```python
export_dir = "./exports"
if not os.path.exists(export_dir):
    os.makedirs(export_dir)
X_user.to_parquet(export_dir + "/X_user.pq")
X_video.to_parquet(export_dir + "/X_video.pq")
y.to_parquet(export_dir + "/y.pq")
```

```python
X_user.to_csv(export_dir + "/X_user.csv")
X_video.to_csv(export_dir + "/X_video.csv")
y.to_csv(export_dir + "/y.csv")
```
