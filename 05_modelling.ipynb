```python
export_dir = "./exports"
left_features_path = export_dir + "/X_user.csv"
right_features_path = export_dir + "/X_video.csv"
labels_path = export_dir + "/y.csv"

left_ds = tf.data.experimental.make_csv_dataset(left_features_path, batch_size=2048, num_epochs=1, shuffle=False)
right_ds = tf.data.experimental.make_csv_dataset(right_features_path, batch_size=2048, num_epochs=1, shuffle=False)
labels_ds = tf.data.experimental.make_csv_dataset(labels_path, batch_size=2048, num_epochs=1, shuffle=False)

def flatten_dict_to_tensor(features_dict):
    tensor_list = [tf.cast(t, tf.float32) for t in features_dict.values()]
    return tf.stack(tensor_list, axis=1)

def zip_and_format(user_ds, video_ds, label_ds):
    dataset = tf.data.Dataset.zip((user_ds, video_ds, label_ds))

    def format_sample(user_feats, video_feats, label_feats):
        user_tensor = flatten_dict_to_tensor(user_feats)
        video_tensor = flatten_dict_to_tensor(video_feats)
        label_tensor = tf.stack(
            [tf.cast(v, tf.float32) for v in label_feats.values()],
            axis=1
        )
        return (user_tensor, video_tensor), label_tensor

    return dataset.map(format_sample).prefetch(tf.data.AUTOTUNE)

# Final dataset
dataset = zip_and_format(left_ds, right_ds, labels_ds)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, Input

# Assume ~70 numeric features per side
NUM_USER_FEATURES = 71
NUM_VIDEO_FEATURES = 73

# User Tower
user_input = Input(shape=(NUM_USER_FEATURES,), name="user_input")
user_tower = layers.Dense(128, activation='relu')(user_input)
user_tower = layers.Dense(64, activation='relu')(user_tower)
user_embedding = layers.Dense(32, activation='relu')(user_tower)

# Video Tower
video_input = Input(shape=(NUM_VIDEO_FEATURES,), name="video_input")
video_tower = layers.Dense(128, activation='relu')(video_input)
video_tower = layers.Dense(64, activation='relu')(video_tower)
video_embedding = layers.Dense(32, activation='relu')(video_tower)

# Merge towers
merged = layers.Concatenate()([user_embedding, video_embedding])
merged = layers.Dense(64, activation='relu')(merged)
merged = layers.Dense(32, activation='relu')(merged)
output = layers.Dense(1, activation='linear')(merged)  # Regression output

# Final model
model = Model(inputs=[user_input, video_input], outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

model.summary()
```

```python
# Use batch and prefetch for performance
#dataset = dataset.batch(1024).prefetch(tf.data.AUTOTUNE)

# Train
model.fit(dataset, epochs=10)
```
